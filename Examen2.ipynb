```python
import pandas as pd
url="/content/daily_weather.csv"
df=pd.read_csv(url)
df.head()
```

```python
df.describe()
```

```python
df['relative_humidity_3pm'].value_counts() > 36
```

```python
# Crear la nueva columna con la condición especificada
df['new_column'] = df['relative_humidity_3pm'].apply(lambda x: 0 if x > 36 else 1)
print(df)
```

```python
df['new_column'].value_counts()
```

```python

df.isnull().any()
```

```python

df_cleaned = df.dropna()

# Mostrar el DataFrame limpio
print(df_cleaned)
```

```python

df_cleaned.isnull().any()
```

```python

df_cleaned
```

```python

df['new_column'].value_counts()
```

```python


tmp=list()
for new_column in df_cleaned['new_column']:
  if new_column== 0:
    tmp.append(1)
  else:
    tmp.append(2)

df_cleaned['target']=tmp
df_cleaned.head(3)
```

```python

df_cleaned['target'].unique()
```

```python
x=df_cleaned.drop(["air_pressure_9am", "air_temp_9am", "avg_wind_direction_9am", "avg_wind_speed_9am", "max_wind_direction_9am", "max_wind_speed_9am", "rain_accumulation_9am", "rain_duration_9am", "relative_humidity_9am", "relative_humidity_3pm", "new_column", 'target'],axis=1)
```

```python
x.head(3)
y=df_cleaned['target']
```

```python

y.head(3)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python

yPredictions = model.predict(x_test)
```

```python
yPredictions
```

```python
#EVALUATIONS METRICS
from sklearn.metrics import accuracy_score
accuracy_score(y_test, yPredictions)
```

```python

# Guardando el modelo
import pickle
with open('model.pkl', 'wb') as file:
    pickle.dump(model, file)
```
